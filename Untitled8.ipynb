Total Revenue Generated

```python
dh['total_price']=dh['UnitPrice']*dh['Quantity']
print('total_price')
print(dh.head())
```

```python
total_revenue=dh['total_price'].sum()
print("Total Revenue:",total_revenue)
```

```python
sns.relplot(x='InvoiceDate',y='total_price',data=dh,kind='line')
plt.title('Total Revenue Over Time')
plt.show()
```

Total orders placed


```python
orders=dh['InvoiceNo'].nunique()
print("Total Orders Placed is",orders)
```

Total products sold

```python
total_products=dh['StockCode'].nunique()
print("Total products :",total_products)
```

Average order value

```python
average_order_value = total_revenue / orders
print("Average Order Value:", average_order_value)
```

Monthly sales

```python
dh['Invoice_Date']=pd.to_datetime(dh['InvoiceDate'])
dh['Month']=dh['Invoice_Date'].dt.month
monthly_revenue=dh.groupby('Month')['total_price'].sum()
print(monthly_revenue)
```

Monthly sales line chart

```python
sns.relplot(x='Month',y='total_price',data=dh,kind='line',height=6,aspect=2)
plt.title('Monthly Revenue')
plt.show()
```

Top 10 products sold

```python
top_10_products=dh.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_10_products)
```

Monthly sales line chart

```python
sns.relplot(x='Description',y='Quantity',kind='line',data=dh,height=6,aspect=2)
plt.title('Top 10 Products')
plt.show()
```

Quarterly sales bar chart

```python
dh['Quarter']=dh['Invoice_Date'].dt.quarter
```

```python
quarter_revenue=dh.groupby('Quarter')['total_price'].sum()
print(quarter_revenue)
```

```python
sns.barplot(x=quarter_revenue.index,y=quarter_revenue.values)
plt.title('Quarterly Revenue')
plt.show()
```

Top products by revenue

```python
top_product_revenue=dh.groupby('Description')['total_price'].sum().sort_values(ascending=False).head(10)
print(top_product_revenue)
```

Least-performing products

```python
least_performing=dh.groupby('Description')['total_price'].sum().sort_values(ascending=True).head(10)
print(least_performing)
```

Number of unique customers

```python
unique_customers=dh['CustomerID'].nunique()
print("Unique Customers:",unique_customers)
print(dh.count())
```

Average spending per customer

```python
average_spending=total_revenue/unique_customers
print("Average Spending Per Customer:",average_spending)
```

Customer order frequency

```python
customer_frequency=dh.groupby('CustomerID')['InvoiceNo'].nunique()
print(customer_frequency)
```

```python
avg_frequency=customer_frequency.mean()
print("Average Frequency:",avg_frequency)
```

```python
sns.relplot(x='CustomerID',y='InvoiceNo',kind='line',data=dh,height=6,aspect=2)
plt.title('Customer Frequency')
plt.show()
```

Revenue by country

```python
revenue_country=dh.groupby('Country')['total_price'].sum().sort_values(ascending=False)
print(revenue_country)
```

```python
sns.barplot(x=top_10_products.index,y=top_10_products.values)
plt.title('Revenue by Country')
plt.show()
```
